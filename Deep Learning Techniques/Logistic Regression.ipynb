```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as py
import tensorflow as tf
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data=load_iris()
print(data)
```

```python
data["target"]
```

```python
data["target_names"]
```

## **Assigning dependant and independant variables**

```python
x=data["data"]
```

```python
y=data["target"]
```

```python
y.shape
```

## **Dataset as Dataframe**

```python
dataset=pd.DataFrame(data=np.concatenate((x,y.reshape(-1,1)),axis=1),columns=["sepal length","sepal width","petal length","petal width","target"])
```

```python
dataset
```

## **Data Split**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,shuffle=True)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## **Building Model**

```python
input=tf.keras.Input(shape=4)
output=tf.keras.layers.Dense(3,activation=tf.keras.activations.sigmoid)(input)
```

```python
model=tf.keras.models.Model(input,output)
```

##**Compiling Model**

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),loss=tf.keras.losses.sparse_categorical_crossentropy,metrics=["accuracy"])
```

```python
model.summary()
```

```python
train=model.fit(x_train,y_train.reshape(-1),validation_data=(x_test,y_test.reshape(-1)),epochs=200)
```

```python
new_data=[5.2,3,4.4,6.5]
```

```python
y_pred= model.predict([new_data])
```

```python
print("Predicted:",y_pred)
print("Predicted class:",np.argmax(y_pred))
```
